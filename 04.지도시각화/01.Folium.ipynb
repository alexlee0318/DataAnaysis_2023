## Folium 지도 시각화

```pip install folium 설치하고 시작```

````python
import folium
folium.__version__
````

- map 종류
    -   -   open street(기본)
    -   -   Stamen Toner

````python
# open street map
map = folium.Map(location=[37.295, 127.045], zoom_start=14) # 시작화면 세팅: 센터 좌표, 줌 정도
folium.Marker(
    location=[37.2949837, 127.0453902], # WGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),    # popup: 클릭했을 때 나타날 내용
    tooltip='경기도청'  # tooltip: 마우스 오버했을 때 나타날 내용
).add_to(map)
map
````

````python
# Stamen Toner 맵
map = folium.Map(location=[37.295, 127.045], zoom_start=14, tiles='Stamen Toner') # 시작화면 세팅: 센터 좌표, 줌 정도
folium.Marker(
    location=[37.2949837, 127.0453902], # WGS84 좌표계, 위도(latitude), 경도(longitude)
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_width=200),    # popup: 클릭했을 때 나타날 내용
    tooltip='경기도청'  # tooltip: 마우스 오버했을 때 나타날 내용
).add_to(map)   # n개 add가능
map
````

- 경기도청, 장안구청, 팔달구청, 권선구청, 영통구청 + Marker 아이콘

````python
# 여러좌표와 Marker 아이콘 꾸미기
map = folium.Map(location=[37.28, 127.02], zoom_start=12)   # 센터값 좌표
folium.Marker(
    location=[37.2949837, 127.0453902],
    popup=folium.Popup('경기도 수원시 영통구 도청로 30', max_sidth=200),
    tooltip='경기도청',
    icon=folium.Icon(
        color='purple', icon_color='white',
        icon='glyphicon glyphicon-thumbs-up',   # 아이콘 제공 사이트
        prefix='glyphicon'
    )
).add_to(map) 
folium.Marker(
    location=[37.30337225, 127.0096118],
    popup=folium.Popup('경기도 수원시 장안구 송원로 101', max_width=200),
    tooltip='장안구청',
    icon=folium.Icon(color='green')
).add_to(map)
folium.Marker(
    location=[37.28256171, 127.0190196],
    popup=folium.Popup('경기도 수원시 팔달구 창룡대로 23', max_width=200),
    tooltip='팔달구청',
    icon=folium.Icon(icon='cloud', color ='red')
).add_to(map)

# 권선구청
folium.Circle(          # Circle
    radius=300, color="crimson", fill=True,
    location=[37.25764489, 126.9719386],
    popup=folium.Popup('경기도 수원시 권선구 호매실로 12', max_width=200),
    tooltip='권선구청'
).add_to(map)

#영통구청
folium.CircleMarker(        # CircleMarker
    radius=50, color="#3186cc",
    fill=True, fill_color="#3186cc",
    location=[37.2596127, 127.0465469],
    popup=folium.Popup('경기도 수원시 영통구 효원로 407', max_width=200),
    tooltip='영통구청'
).add_to(map)

title_html = '<h3 align="center" style="font-size:20px">수원 소재 주요 관공서</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
````
