````python
import pandas as pd
import numpy as np
````

```d1, d2의 벤다이어그램```

````python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
````

````python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
````

````python
# Inner join (교집합): 같은 컬럼명을 기준으로, 공통된 데이터만 합성
pd.merge(df1, df2)
````

````python
# Full outer join
pd.merge(df1, df2, how='outer')
````

````python
# Lefter outer join
pd.merge(df1, df2, how='left')
````

````python
# Right outer join
pd.merge(df1, df2, how = 'right')
````

- join method

````python
# 인덱스로 join하기
df1.set_index('고객번호', inplace=True)
df2.set_index('고객번호', inplace=True)
````

````python
df1.join(df2)   # left outer join
````

````python
pd.merge(df1, df2, left_index=True, right_index=True)
````

연습 문제 4.6.1

두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다.
각각 5 x 5 이상의 크기를 가진다.
공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다.

````python
df3 = pd.DataFrame(np.random.randint(100, 110, size=(5,5)), columns =list('ABCDE'))
df3
````

````python
df4 = pd.DataFrame(np.random.randint(100, 110, size=(5,5)), columns =list('ABXYZ')) # AB열 공통

df4
````

````python
# 실험해보기 - 두 표의 A,B값 같게 만들기
df4.B[1] = 101
df4
````

````python
pd.merge(df3, df4, on = 'A')
````

````python
pd.merge(df3, df4, on = 'B')
````

````python
pd.merge(df3, df4, on = ['A','B'])
````

```
연습 문제 2
어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 실적 정보는 “매출”, “비용”, “이익” 으로 이루어진다.
(이익 = 매출 - 비용).
또한 1년간의 총 실적을 마지막 행으로 덧붙인다.
```

````python
df5 = pd.DataFrame(np.random.randint(500, 1000, size=(6, 2)) * 1000,
                    columns=['매출','비용'], index=list('123456'))
df5['이익'] = df5['매출'] - df5['비용']
df5
````

````python
df6 = pd.DataFrame(np.random.randint(500, 1000, size=(6, 2)) * 1000,
                    columns=['매출','비용'],
                    index=['7','8','9','10','11','12'])
df6['이익'] = df6['매출'] - df6['비용']
df6
````

````python
df7 = pd.concat([df5, df6])
df7
````

````python
# 행추가하기 _ 연간 합계 산출: 매출/비용/이익 각각의 연간합계
df7.loc['합계'] = df7.sum()
df7.tail()
````
