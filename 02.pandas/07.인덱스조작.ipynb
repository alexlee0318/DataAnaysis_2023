- 인덱스 설정 및 제거

````python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,
                   columns=["C1", "C2", "C3", "C4"])
df1
````

set_index 메서드로 특정한 열을 인덱스로 설정할 수 있다. 이 때 기존의 인덱스는 없어진다.

````python
df2 = df1.set_index('C1')   # df1.set_index('C1', inplace=True)
df2
````

````python
df1.index, df2.index
````

````python
df3 = df2.set_index('C2')
df3
````

````python
df2.reset_index(inplace=True)   # 다시 되돌리기
df2
````

```
연습 문제 4.5.1

5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다.

학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후, df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. reset_index 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.
```

````python
df_score1 = pd.DataFrame(np.random.randint(60, 101, size=(5,3)),
                         columns=['국어','영어','수학'])
df_score1
````

````python
df_score1.index = ['피카츄','또가스','파이리','꼬북이','야도란']
df_score1.index.name = '포켓몬'
df_score1
````

````python
df_score2 = df_score1.reset_index()
df_score2
````

````python
# 위 식을
df_score2.set_index('포켓몬', inplace=True)
df_score2
````

- 다중 인덱스 : 만들 일은 x, 데이터에 접근하는 방식들 알기

```인덱스가 2개일 때```

````python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])  # 인덱스 2개임
df3
````

````python
df3.columns.names = ['Cidx1', 'Cidx2']  # 이름을 지정해봄
df3
````

````python
# 엑세스 하기 _ 상위 인덱스부터 좌표 찍어줘야 함
df3['A','C1']
````

```인덱스 2개 + 행 2개일 때```

````python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
````

````python
# 0.41값 인덱싱 하기_1
df4['A','D']['M','id_3']
````

````python
# 0.41값 인덱싱 하기_2
df4.iloc[2, 1], df4.loc[('M','id_3'),('A','D')]
````
