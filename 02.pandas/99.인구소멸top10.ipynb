22.12기준 인구집계현황 데이터 - 인구소멸 위기지역 Top10 구하기

- 기초자치단체(시, 군)
특례시: 수원, 고양, 용인(구)
행정구: 성남, 안산, 안양(구)

```python
df = pd.read_csv('data/주민등록인구집계현황.csv',encoding='euc=kr')
df.head(1)
```

```python
# 카테고리 정리/ 셀렉
df = df[['행정구역구분명','행정구역명','20~29세 (여)','30~39세 (여)','60~69세','70~79세','80~89세','90~99세','100세 이상']]
df.set_index('행정구역구분명', inplace=True)
df.drop('읍면동', inplace=True)

df.head()
```

```python
# 노령화 비율 구하기
df['가임기여성수'] = df['20~29세 (여)'] + df['30~39세 (여)']
df['노령인구'] = df.loc[:, '60~69세':'100세 이상'].sum(axis=1)

df['노령화비율'] = df['가임기여성수'] / df['노령인구']
df.head(2)
```

```python
df=df[['행정구역명','가임기여성수','노령인구','노령화비율']]
df.head(2)
```

```python
# 도, 특례시, 광역시 데이터 제거하기(~시로 끝나는 데이터만 필터링 .str.endswith() )

# df['행정구역'] = df['행정구역명'].str.split(" ", expand=True)         # # 행정구역명 str으로 만들고, 분할된 개별 리스트를 데이터 프레임으로 전환
# df[df['행정구역'].endswith("시")]
df['행정구역'] = df['행정구역명'].str.findall('\w+도')

# df.drop(index= df[df['행정구역']])

df.head()
```
