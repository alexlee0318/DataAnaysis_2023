````python
np.random.seed = 2023
df = pd.DataFrame(np.random.randint(40, 101, size= (4, 5)),
                index=list('abcd'), columns=list('VWXYZ'))
df
````

```
연습 문제 4.4.4
타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라.
category1 카테고리는 다음과 같이 정의된다.
- 20살이 넘으면 성별을 그대로 사용한다.
- 20살 미만이면 성별에 관계없이 “child”라고 한다.
```

````python
# LV.2_ apply로 풀기
titanic['categry1'] = titanic.apply(lambda x: x.sex if x.age >= 20 else 'child', axis=1)    #axis를 안주면 같은 행에 대해 계산하게 됨
titanic
# titanic > lambda 순이므로 DF > col 
````

````python
# LV.1_ for문으로 풀기
cat2 = []
for i in titanic.index: # range가 필요없음. titanic의 index 자체가 i가 됨
    if titanic.age[i] > 20:
        cat2.append(titanic.sex[i]) #single value의 i
    else:
        cat2.append('child')

titanic['cat2'] = cat2
titanic.tail(5)
````

- 실수값을 카테고리 값으로 변환
    -   cut: 실수 값의 경계선을 지정하는 경우
    -   qcut: 갯수가 똑같은 구간으로 나누는 경우

````python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
````

````python
bins = [1, 20, 30, 50, 70, 100]     # 기준 입력_  1 < 미성년자 <= 20
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels=labels)
cats
````

````python
df5 = pd.DataFrame(ages, columns=["ages"])
df5["age_cat"] = pd.cut(df5.ages, bins, labels=labels)
df5
````

````python
df5['one_over_n'] = pd.qcut(ages, 4, labels=['Q1', 'Q2','Q3','Q4']) # 각 Q값의 수가 3개가 되도록 균등분배
df5
````
