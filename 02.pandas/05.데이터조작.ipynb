- 정렬
    -       *   sort_values() 와 sort_index()

````python
np.random.seed = 2023
df = pd.DataFrame(np.random.randint(40, 101, size= (4, 5)),
                index=list('abcd'), columns=list('VWXYZ'))
df['V']['C'] = 66
df
````

````python
df.sort_values(by='V', ascending=False)
````

````python
# V열에 동일한 값이있으면 동일한 값은 W열을 보고 정렬하기
df2 = df.sort_values(by=['V','W'], ascending=False)
df2
````

````python
df2.sort_index()
````

``` 연습문제 4.4.3 타이타닉호 승객의 나이를 조건에 따라 구하기```

````python
titanic
````

````python
# 1)타이타닉호 승객의 평균 나이를 구하라.
titanic.age.mean().round(2)
````

````python
# 2) 여성 승객의 평균 나이를 구하라 - step 1. 여성승객의 DF 추출

ft = titanic[titanic.sex == 'female']
ft.head(3)
````

````python
# 2) 여성 승객의 평균 나이를 구하라 - step 2. 평균나이 추출
ft.age.mean().round(2)
````

````python
titanic[titanic.sex == 'female'].age.mean().round(2) # []안에는 특정 조건
````

````python
# 3) LV2.1등실 선실의 여성 승객의 평균 나이를 구하라. 
titanic[(titanic.pclass ==1) & (titanic.sex =='female')]['age'].mean().round(2)   # titanic[DF][col].mean
````

````python
#  3) LV1. 1등실 선실의 여성 승객의 평균 나이를 구하라.(LV2를 쉽게 풀어 쓴 식)
dt1 = titanic[titanic.pclass == 1]
dt2 = dt1[dt1.sex =='female']  # titanic을 쓰지않고 다른 변수로(dt1) 바꿔줘야 함(data frame이 맞지 않기때문)
dt2.age.mean().round(2)
````

```1. 타이타닉호의 승객 중 나이 20살을 기준으로 성인(adult)과 미성년자(child)를 구별하는 라벨 열을 만들기
```

``` 타이타닉호 승객 중 나이 20살을 기준으로 성인(adult)과 미성년자(child)를 구별하는 라벨 열을 만들어라```

````python
titanic["adult/child"] = titanic.age.apply(lambda x: "adult" if x >= 20 else "child")
titanic.tail()  # tail() = 아래부터 5개 보여줌
````

``` 연습 문제 4.4.4
타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라.
category1 카테고리는 다음과 같이 정의된다.

- 20살이 넘으면 성별을 그대로 사용한다.
- 20살 미만이면 성별에 관계없이 “child”라고 한다.
```
