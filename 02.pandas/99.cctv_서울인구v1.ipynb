1. cctv 데이터를 다음과 같이 만들어 cctv.csv 파일에 저장하세요.

``` 
            구별    CCTV댓수  최근증가율

            1 종로구 1715         20.69

            2 중구    2447         78.98

            3 용산구 2611          78.64

                    ...   

````python
filename = 'data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
df = pd.read_csv(filename, encoding='euc-kr', thousands=',')
````

````python
#1) 첫번째 행인 계는 삭제한다.
df.drop(0, inplace=True)    # drop([col,row])

#2) 최근증가율 구하고 열 만들기
df['최근증가율'] = (((df['2020년']+df['2021년']+df['2022년']) /
                    (df['2013년 이전 설치된 CCTV']+df['2013년']+df['2014년']+df['2015년']+df['2016년']+df['2017년']+df['2018년']))
                    * 100).round(2)
df.head(2)
````

````python
#3)구별은 기존의 구분 열을 CCTV댓수는 기존의 총계 열을 이름만 변경한 것이다.
df.rename(columns={
    '구분': '구별',
    '총계': 'CCTV댓수',
    '최근증가율': '최근증가율'
}, inplace=True)


cctv = df[['구별','CCTV댓수','최근증가율']]
cctv.head()
````

````python
# cctv 파일이름 변경/저장
cctv.to_csv('data/cctv.csv', index=False)

# cctv.csv 읽기
cctv = pd.read_csv('data/cctv.csv')
cctv.head()
````

2. 서울시 인구 데이터를 다음과 같이 만들어 pop.csv 파일에 저장하세요.
```
  구별     인구수  내국인  외국인  고령자  외국인비율  고령자비율
1 종로구 153789 144683  9106    27818   5.92           18.09
2 중구    131787 122499  9288    24392   7.05           18.51
3 용산구 237285 222953  14332   39070   6.04           16.47
```

````python
# 파일불러오기
df1 = pd.read_csv('data/서울시인구.txt', sep='\t', skiprows=2, thousands=',')  
df1.head()
````

````python
# '합계' 행 삭제
df1 = df1.drop(index=0,axis=0)
# del pop1['기간']

df1.head()
````

````python
# coulmn명 변경
df1.rename(columns={
    '자치구': '구별',
    '계': '인구수',
    '계.1': '내국인',
    '계.2': '외국인',
    '65세이상고령자': '고령자',
}, inplace=True)

# DF 수정
pop = df1[['구별','인구수','내국인','외국인','고령자']]   # pop = df1.iloc[:, [1, 3, 6, 9, 13] ].copy(True)
pop.head()
````

````python
# '외국인비율' 열 추가(외국인 수/ 총 인구수 *100)
# '고령자비율' 열 추가(65세 이상 인구 수/ 총 인구수 *100)
# pop.apply(lambda x: round(x['외국인(고령자)']/x['인구수'] * 100, 2), axis=1)
import warnings
warnings.filterwarnings('ignore')

pop['외국인비율'] = (((pop['외국인'])/(pop['인구수']))*100).round(2)
pop['고령자비율'] = (((pop['고령자'])/(pop['인구수']))*100).round(2)
pop.head(2)
````

````python
# pop 파일 이름 변경/저장
pop.to_csv('data/pop.csv', index=False)

# pop 파일 읽기
pop = pd.read_csv('data/pop.csv')
pop
````
