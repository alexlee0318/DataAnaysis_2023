### 2. 여러가지 플롯

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False      # minus오류 없애는
````

1. Bar Chart 

````python
# 안좋은 예
y, x = [2, 3, 1],[0, 1, 2]
plt.title('Bar Chart')
plt.bar(x, y)
plt.xticks(x, ['가','나','다']), plt.yticks(sorted(y))
plt.xlabel('가나다'), plt.ylabel('빈도수')
````

````python
# 더 나은 예 
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar chart')
plt.yticks([0, 0.2, 0.4, 0.6, 0.8]), plt.xticks([]), plt.grid() # grid 정돈- x축에는 안넣겠다
plt.show()
````

````python
s.plot(kind='bar')
plt.show()
````

````python
# 더 나은 예 
df = pd.DataFrame(np.random.rand(20).reshape(10,2),
                  index=list('abcdefghij'), columns=['M','N'])
plt.bar(df.index, df.M.values)
plt.title('Bar chart')
plt.show()
````

- 타이타닉 데이터로 누운 차트 만들기

````python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
````

````python
s = titanic['class'].value_counts()
plt.figure(figsize=(3,1))
plt.barh(s.index, s.values) # barh에는 x, y 밸류 주면 됨
plt.title('타이타닉호의 선실별 승객수')
plt.show()
````

#### 2. Pie Chart

````python
# 기본모양
ratio = [10, 15, 20, 25]        # 합이 100이 아니어도 됨
labels = ['감','귤','배','밤 ']
plt.figure(figsize=(5,2))
plt.pie(ratio, labels=labels, autopct='%.1f%%') #소숫점 1째자리, %기호 쓰겠음
plt.title('파이 차트')
plt.show()
````

````python
colors = ['yellowgreen', 'gold','lightskyblue','lightcoral']
explode = (0, 0.1, 0, 0)
plt.figure(figsize=(5,2))
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=colors, explode=explode,
        startangle=90, shadow=True)
plt.title('파이차트')
````

4. 산점도(Scatter plot) : x, y의 관계를 눈으로 보기
    -   2차원 데이터 즉, 두 개의 실수 데이터 집합의 상관관계를 살펴보려면 scatter 명령으로 스캐터 플롯을 그린다. 스캐터 플롯의 점 하나의 위치는 데이터 하나의 x, y 값이다.

````python
np.random.seed(2023)
X, Y = np.random.normal(2, 2, 100), np.random.randn(100)    # 평균2, 표준편차 2 인 숫자 100개
plt.figure(figsize=(4,2))
plt.scatter(X, Y)
plt.title('산점도_X,Y관계')
plt.show()
````

````python
colors = np.random.rand(100)    # 색상값 범위
plt.figure(figsize=(4,2))
plt.scatter(X, Y, c=colors, alpha=0.5) # 색상, 투명도
plt.title('색상을 준 산점도')
plt.show()
````

````python
size = 500 * np.random.rand(100)    # 크기
plt.figure(figsize=(5,3))
plt.scatter(X, Y, s=size, c=colors, alpha=0.5) # 크기, 색상, 투명도
plt.title('색상을 준 산점도')
plt.show()
````
