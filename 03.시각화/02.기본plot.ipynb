### 1. Matplotlib 소개

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False      # minus오류 없애는
```

- 라인 플롯

```python
plt.plot([0,1,2,3],[0,1,4,9])     # X 생략가능(어차피 y 개수만큼 0부터~되므로)
plt.title('라인 플롯'); plt.grid()  # 한 줄에 여러 식 가능(;또는 ,로 구분)
```

- 1.1 스타일 지정
- https://datascienceschool.net/01%20python/05.01%20%EC%8B%9C%EA%B0%81%ED%99%94%20%ED%8C%A8%ED%82%A4%EC%A7%80%20%EB%A7%B7%ED%94%8C%EB%A1%AF%EB%A6%AC%EB%B8%8C%20%EC%86%8C%EA%B0%9C.html

```python
plt.plot([0,1,2,3],[0,1,4,9], 'rs--')    # color, marker, line style
plt.title("'rs--' 스타일의 플롯")
plt.grid()
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'mo-.')    # color, marker, line style
plt.plot([0,1,2,3],[9,4,1,0], 'g^:')
plt.title("mo-. 와 g^: 스타일의 플롯")
plt.grid()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'bo--',
         linewidth=5, markersize=15, markeredgecolor="g", markeredgewidth=5, markerfacecolor="r")  # lw, ms, mec, mew, mfc,로 줄임 가능
plt.title("스타일 적용 예"), plt.grid()
```

- 그림 범위

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("x축, y축의 범위 설정")
plt.xlim(0, 50), plt.ylim(-10, 30), plt.grid()
```

- 곡선 그림

```python
X = np.linspace(0, 2*np.pi, 500)    # X = np.arange(0, 2*np.pi+0.01, 0.01) 같음
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sine, Cosine 그래프'), plt.grid()
```

- 1.3 Tick 설정 ( 격자를 원하는 자리에 설정)

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])     # 5군데
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('LaTeX, 문자열로 tick label 정의')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])  # LaTeX
plt.yticks([-1, 0, 1]), plt.grid()
plt.show()
```

- 범례(legend)

```python
plt.plot(X, Ysin, label='Sine')   # Sine라벨
plt.plot(X, Ycos, label='Cosine')   # Cosine라벨
plt.title('범례 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc=0)   # best position
plt.show()
```

```python
plt.plot(X, Ysin, label='Sine') # Sine라벨
plt.plot(X, Ycos, label='Cosine')   # Cosine라벨
plt.title('범례 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc='9')   # loc='upper center position' 과 동일
plt.show()
```

- X축, Y축 Label

```python
plt.plot(X, Ysin, label='Sine')
plt.plot(X, Ycos, label='Cosine')
plt.title('X축, Y축 Label 표시')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')  # 축 라벨링
plt.show()
```

- 그림의 크기 지정

```python
plt.figure(figsize=(5,2))   # 크기 지정
plt.plot(X, Ysin, label='Sine')
plt.plot(X, Ycos, label='Cosine')
plt.title('그림 크기 지정(5,2)')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')
plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(5,2))
plt.plot(X, Ysin, label='Sine')
plt.plot(X, Ycos, label='Cosine')
plt.title('그림 저장', fontsize=15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.yticks([-1, 0, 1]), plt.grid()
plt.legend(loc=0)
plt.xlabel('시간'), plt.ylabel('진폭')
plt.savefig('data/삼각함수.png')    # 삼각함수.png 파일로 저장
```
